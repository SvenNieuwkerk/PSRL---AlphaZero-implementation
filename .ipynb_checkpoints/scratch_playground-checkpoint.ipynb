# Imports

```python
import gymnasium as gym
from acorl.envs.seeker.seeker import SeekerEnv, SeekerEnvConfig
from acorl.env_wrapper.tianshou.env_factory import InputSetComputationEnvFactory
from acorl.envs.constraints.seeker import SeekerInputSetPolytopeCalculator

import numpy as np
import matplotlib.pyplot as plt
```

## Setting up a basic environment

```python
#environment configuration (dataclass of some sorts; pydantic flavor)
env_config = SeekerEnvConfig(randomize=False, num_obstacles=1)
#create actual environment object from config
env = SeekerEnv(**env_config.model_dump(exclude={"id"}))
#
obs, info = env.reset()
obs, info
```

## Plotting helper functions

```python
import numpy as np
import matplotlib.pyplot as plt

def decode_obs(obs):
    obs = np.asarray(obs)
    agent = obs[0:2]
    goal = obs[2:4]
    obstacles_flat = obs[4:]
    num_obstacles = len(obstacles_flat)//3
    obstacles = obstacles_flat.reshape(num_obstacles, 3)  # (x, y, r) per obstacle
    return agent, goal, obstacles

def plot_seeker_obs(obs, info, title=None):
    agent, goal, obstacles = decode_obs(obs)
    L = info.get("boundary_size", 10)

    fig, ax = plt.subplots(figsize=(6, 6))

    # Obstacles
    for (ox, oy, r) in obstacles:
        circ = plt.Circle((ox, oy), r, color="red", alpha=0.3)
        ax.add_patch(circ)

    # Goal
    ax.scatter(goal[0], goal[1], color="green", s=80, label="Goal")
    # (optional: draw goal radius if you know it somewhere else)

    # Agent
    ax.scatter(agent[0], agent[1], color="blue", s=80, label="Seeker")

    ax.set_xlim(-L, L)
    ax.set_ylim(-L, L)
    ax.set_aspect("equal", adjustable="box")
    ax.legend()
    if title:
        ax.set_title(title)
    plt.show()
```

```python
# Make sure you know how many obstacles this env actually has:
num_obstacles = (len(obs) - 4) // 3
# also just env._num_obstacles
print("num_obstacles =", num_obstacles)

plot_seeker_obs(obs, info, num_obstacles)
```

## Take an action

```python
# Define a simple action: move slightly to the right
action = np.array([0.5, 0.0], dtype=np.float64)

next_obs, reward, terminated, truncated, info2 = env.step(action)

print("reward:", reward)
print("terminated:", terminated, "truncated:", truncated)
plot_seeker_obs(next_obs, info2, title="After one step")
```

```python
# move to the right a bit more
action = np.array([1.0, 0.0], dtype=np.float64)
# action = np.array([2.0, 0.0], dtype=np.float64) would give out of bounds ValueError since 2 > 1

next_obs, reward, terminated, truncated, info2 = env.step(action)
next_obs, reward, terminated, truncated, info2 = env.step(action)
next_obs, reward, terminated, truncated, info2 = env.step(action)

print("reward:", reward)
print("terminated:", terminated, "truncated:", truncated)
plot_seeker_obs(next_obs, info2, title="After one step")
```

```python
obs, info = env.reset()
done = False
step_i = 0

plot_seeker_obs(obs, info, title=f"Step {step_i}")

while not done and step_i < 20:
    # Example: always move towards the goal center (very naive)
    agent, goal, obstacles = decode_obs(obs)
    direction = goal - agent
    if np.linalg.norm(direction) > 0:
        direction = direction / np.linalg.norm(direction)  # unit vector

    # Scale by max action_size (here 1.0)
    action = direction * 0.5  # half-strength step

    next_obs, reward, terminated, truncated, info = env.step(action)
    step_i += 1
    print(f"Step {step_i}: reward={reward}, terminated={terminated}, truncated={truncated}")
    plot_seeker_obs(next_obs, info, title=f"Step {step_i}")

    obs = next_obs
    done = terminated or truncated
```
